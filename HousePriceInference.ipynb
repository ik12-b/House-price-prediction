```python
import joblib
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
model = joblib.load('/content/drive/MyDrive/model/housepricepred.pkl')
columns = joblib.load('/content/drive/MyDrive/pre-processing/columns.pkl')
def predict(Address,Zip,Area,Room,Lon,Lat):
  data = {'Address':Address,'Zip':Zip,'Area':Area,'Room':Room,'Lon':Lon,'Lat':Lat}
  data = pd.DataFrame(data,index=[0])
  data = pd.get_dummies(data,columns=['Address','Zip'],drop_first=False)
  data = data.reindex(columns=columns)
  output = model.predict(data)
  output = np.exp(output)[0]
  output = f"price :,{int(output/1000)}k"
  return output
predict("AZaaiersweg 11 Amsterdam","1033 KM",55,4,40.453543,60.543267)


```
